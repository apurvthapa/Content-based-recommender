## Text to Vectors (Vectorisation)

```python
# Using bag of words technique
# in it we combine the tags of all the movies, then make a record of 5000 words that are most repeated.
# then from every movie tags check how many times the most repeated words are repeated for that movie.(this is a vector)
# the movies with most similar repeated words will be recommended.
# Note- will not consider stop words like [are,to,and,if,from,a] 
# (words that are used for sentence formation but not for its meaning).
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names()  #5000 most frequently used words
```

```python
len(cv.get_feature_names())
```

```python
# just by manually looking at words we have seen that words like activity, activities or actor, actors are taken 
# as different words whereas it means the same.
# We use stemming technique. which converts words like loving , loves to love.
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
ps.stem('actors')
```

```python
ps.stem('dancing')
```

```python
def stem(txt):
    y=[]
    for i in txt.split():
        y.append(ps.stem(i))
    return ' '.join(y)
```

```python
stem('love i loving actors actor activities activity')
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

# create a function to recommend the movie

```python
new_df[new_df['title']=='Batman Begins'].index[0]  #get the index of the movie
```

```python
similarity[119] #distance of batman begins with all the other movies.
```

```python
# but the problem with above code is that by doing so we have lost the index number of the movie.
# eg we know that movie at 119 has 1 distance with itself but we do not know the index of the next movie.
# Hence we apply enumerate function. So that the indexes can be fixed
```

```python
list(enumerate(similarity[119])) # now every distance has the index number with it in the tuple.
```

```python
sorted(list(enumerate(similarity[119])),reverse=True) # though the tuples are sorted but are on the basis of index.
```

```python
sorted(list(enumerate(similarity[119])),reverse=True,key= lambda x:x[1])
```

```python
# hence the top 5 similar movie with 119 are 65, 1363,1362,3, 3297.
```

```python
sorted(list(enumerate(similarity[119])),reverse=True,key= lambda x:x[1])[1:6]
```

```python
new_df.iloc[1363].title
```

```python
def recommend (movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movie_list=sorted(list(enumerate(distances)),reverse=True,key= lambda x:x[1])[1:6]
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```

```python
recommend('Avatar')
```
